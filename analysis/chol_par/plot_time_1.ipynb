```python
import numpy as np
import math  as m
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
rc('text', usetex=True)
mpl.rcParams['axes.linewidth'] = 2
mpl.rcParams.update({'figure.autolayout': True})
```

```python
INPUT1 = open("../1/chol_par_10.txt", "r")
lines1 = INPUT1.readlines()
N1     = len(lines1)
t1     = np.zeros(N1)
a1     = np.zeros(N1)

INPUT2 = open("../1/chol_par_20.txt", "r")
lines2 = INPUT2.readlines()
N2     = len(lines2)
t2     = np.zeros(N2)
a2     = np.zeros(N2)

INPUT3 = open("../1/chol_par_30.txt", "r")
lines3 = INPUT3.readlines()
N3     = len(lines3)
t3     = np.zeros(N3)
a3     = np.zeros(N3)

INPUT4 = open("../1/chol_par_40.txt", "r")
lines4 = INPUT4.readlines()
N4     = len(lines4)
t4     = np.zeros(N4)
a4     = np.zeros(N4)

INPUT5 = open("../1/chol_par_50.txt", "r")
lines5 = INPUT5.readlines()
N5     = len(lines5)
t5     = np.zeros(N5)
a5     = np.zeros(N5)

for i in range(0, N1):
    line0     = lines1[0].split()
    line      = lines1[i].split()
    t1[i]     = float(line[0]) - float(line0[0])
    a1[i]     = (1 - float(line[1])) / float(line[1])

for i in range(0, N2):
    line0     = lines2[0].split()
    line      = lines2[i].split()
    t2[i]     = float(line[0]) - float(line0[0])
    a2[i]     = (1 - float(line[1])) / float(line[1])

for i in range(0, N3):
    line0     = lines3[0].split()
    line      = lines3[i].split()
    t3[i]     = float(line[0]) - float(line0[0])
    a3[i]     = (1 - float(line[1])) / float(line[1])

for i in range(0, N4):
    line0     = lines4[0].split()
    line      = lines4[i].split()
    t4[i]     = float(line[0]) - float(line0[0])
    a4[i]     = (1 - float(line[1])) / float(line[1])

for i in range(0, N5):
    line0     = lines5[0].split()
    line      = lines5[i].split()
    t5[i]     = float(line[0]) - float(line0[0])
    a5[i]     = (1 - float(line[1])) / float(line[1])
```

```python
        
plt.plot(t1 * 5, a1, color ='red',linewidth = 3, label = '$\mathrm{10}\, \%$')
plt.plot(t2 * 5, a2, color ='blue',linewidth = 3, label = '$\mathrm{20}\, \%$')
plt.plot(t3 * 5, a3, color ='orange',linewidth = 3, label = '$\mathrm{30}\, \%$')
plt.plot(t4 * 5, a4, color ='cyan',linewidth = 3, label = '$\mathrm{40}\, \%$')
plt.plot(t5 * 5, a5, color ='purple',linewidth = 3, label = '$\mathrm{50}\, \%$')

plt.ylabel('$ \mathrm{Cholesterol \ Partitioning}$', fontsize=25)
plt.xlabel('$\mathrm{Time} \ (\mathrm{ns})$',fontsize=25)
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
plt.xlim((0, 1000))
plt.ylim((-0.005, 0.06))
plt.yticks(np.arange(0, 0.061, step=0.02))
#plt.xticks(np.arange(10, 51, step=10))
plt.legend(fontsize=15, ncol=3, borderpad=0.2)
plt.savefig("cholestrol_partitioning_SM86_time-1.png")
plt.savefig("cholestrol_partitioning_SM86_time-1.pdf")

# plt.close()
```
