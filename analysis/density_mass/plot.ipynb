```python
import numpy as np
import math  as m
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
rc('text', usetex=True)
mpl.rcParams['axes.linewidth'] = 2
mpl.rcParams.update({'figure.autolayout': True})

```

```python
z1_1    = []
z2_1    = []
z1_2    = []
z2_2    = []
z1_3    = []
z2_3    = []
z1      = np.zeros(100)
z2      = np.zeros(100)
z1_err  = np.zeros(100)
z2_err  = np.zeros(100)
R     = []
x=[]
INPUT1   =  open("../1/density-0.txt" , "r")
INPUT2   =  open("../2/density-0.txt" , "r")
INPUT3   =  open("../3/density-0.txt" , "r")

lines1 = INPUT1.readlines()
lines2 = INPUT2.readlines()
lines3 = INPUT3.readlines()

for j in range(0, int(len(lines1))):
    line = lines1[j].split()
    R.append(float(line[0]))
    z1_1.append(float(line[1]))
    z2_1.append(float(line[2]))

for j in range(0, int(len(lines2))):
    line = lines2[j].split()
    z1_2.append(float(line[1]))
    z2_2.append(float(line[2]))

for j in range(0, int(len(lines3))):
    line = lines3[j].split()
    z1_3.append(float(line[1]))
    z2_3.append(float(line[2]))

for i in range(0, len(z1_1)):
    x.append(i)
#     print (i)
    z1[i] = (z1_1[i] + z1_2[i] + z1_3[i]) / 3.0
    z2[i] = (z2_1[i] + z2_2[i] + z2_3[i]) / 3.0
    z1_err[i] = (((z1_1[i]-z1[i])**2 + (z1_2[i]-z1[i])**2 + (z1_3[i]-z1[i])**2)/3.0)**0.5/(3**0.5)
    z2_err[i] = (((z2_1[i]-z2[i])**2 + (z2_2[i]-z2[i])**2 + (z2_3[i]-z2[i])**2)/3.0)*0.5/(3**0.5)
```

```python

#plt.plot(z1 / np.max(z2), R, color='cyan', linewidth=4, label='$\mathrm{N-Ionizable \ Lipids}$')
plt.plot(z2 / np.max(z1), R, color='cyan', linewidth=4, label='$\mathrm{N-Neutral}$')
plt.plot(z1 / np.max(z1), R, color='red', linewidth=4, label='$\mathrm{N-Protonated}$')
#plt.plot(z3 / np.max(z2), R, color='orange', linewidth=4, label='$\mathrm{Na ^ {+}}$')
plt.xlabel('$\mathrm{Scaled \ mass\ density}$', fontsize=25)
plt.ylabel('$z \ (\mathrm{\AA})$',fontsize=25)
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
plt.ylim((-60, 60))
plt.xlim((0, 1))
plt.yticks(np.arange(-80, 80.1, step=40))
# plt.xticks(np.arange(-40, 50, step=20))
plt.legend(fontsize=15)
plt.savefig("density_SM86_10.pdf")
```
