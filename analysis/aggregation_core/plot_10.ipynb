```python
import numpy as np
import math  as m
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
rc('text', usetex=True)
mpl.rcParams['axes.linewidth'] = 2
mpl.rcParams.update({'figure.autolayout': True})

INPUT1   =  open("coor-p-0.txt" , "r")
INPUT2   =  open("coor-n-0.txt" , "r")
```

```python
lines1    = INPUT1.readlines()
lines2    = INPUT2.readlines()

x_p_u = np.array([])
y_p_u = np.array([])
z_p_u = np.array([])
x_n = np.array([])
y_n = np.array([])
z_n = np.array([])
x_p_l = np.array([])
y_p_l = np.array([])
z_p_l = np.array([])

for i in range(0, len(lines1)):
    line = lines1[i].split()
    if float(line[3]) > 0:
        x_p_u = (np.append(x_p_u, float(line[1])))
        y_p_u = (np.append(y_p_u, float(line[2])))
        z_p_u = (np.append(z_p_u, float(line[3])))
    if float(line[3]) < 0:
        x_p_l = (np.append(x_p_l, float(line[1])))
        y_p_l = (np.append(y_p_l, float(line[2])))
        z_p_l = (np.append(z_p_l, float(line[3])))

for i in range(0, len(lines2)):
    line = lines2[i].split()
    
    x_n = (np.append(x_n, float(line[1])))
    y_n = (np.append(y_n, float(line[2])))
    z_n = (np.append(z_n, float(line[3])))

print ('done!')
```

```python
plt.scatter(x_p_u, z_p_u,  c = z_p_u, cmap = 'Oranges', s = 1)
plt.scatter(x_p_l, z_p_l,  c = z_p_l, cmap = 'Oranges_r', s = 1)

plt.plot(x_n, z_n, '.', color = 'cyan', markersize = 1)
plt.ylabel('$z \ (\mathrm{\AA})$', fontsize=25)
plt.xlabel('$x \ (\mathrm{\AA})$', fontsize=25)
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
plt.xlim((-60, 60))
plt.ylim((-60, 60))
plt.xticks(np.arange(-60, 60.1, step=20))
plt.yticks(np.arange(-60, 60.1, step=20))
plt.savefig('crowding-10-1.pdf')
plt.savefig('crowding-10-1.png')
```
